# Exercice 4: Développer une stratégie en utilisant applymap() pour obtenir la série qui donne la semaine correspondant à la date d’achat

```python
df5 = pd.DataFrame({'date_achat':["2019-02-14", "2018-07-13", "2019-12-11"], 'qte':["175", "180", "1.95"]})
```

```python
df5['date_achat_relle'] = pd.to_datetime(df5['date_achat'], format='%Y-%m-%d')
```

```python
df5
```

```python
df5['semaine'] = df5[['date_achat_relle']].applymap(lambda dt:dt.week if not pd.isnull(dt.week) 
                                                            else 0)
```

```python
df5
```

# Exercice 6: On dispose de deux dataframes portant sur des vins blancs et rouges. On veut fusionner les deux dataframes

```python
# 1-Lecture des dataset
wine_quality_red = pd.read_csv('./data/winequality-red.csv', delimiter=';')
wine_quality_white = pd.read_csv('./data/winequality-white.csv', delimiter=';')
```

```python
wine_quality_red
```

```python
wine_quality_white
```

```python
# 2-Ajouter une serie "wine_type" qui permettra de différencier le vin dans le dataset final. Les valeurs à utiliser ici sont : 
# "red" et "white"
wine_quality_red['wine_type'] = 'red'
wine_quality_white['wine_type'] = 'white'
```

```python
# 3-Sur la base de la serie "quality", ajouter une nouvelle serie "wine_quality" qui donne la qualité categorielle du vin selon 
# la regle
# On peut utiliser ce dictionnaire pour faire un map
wine_quality = {3:'low quality', 4:'low quality', 5:'low quality', 
                6:'Medium quality', 7:'Medium quality', 
                8:'High quality', 9:'High quality'}
```

```python
# wine_quality_red['wine_quality'] = wine_quality_red['quality'].map(wine_quality)
wine_quality_red['wine_quality'] = wine_quality_red['quality'].apply(lambda x: 'low quality' if x<=5 
                                                                     else 'Medium quality' if x<=7 
                                                                     else 'High quality')
```

```python
# Categories
wine_categorie = pd.Categorical(wine_quality_red['wine_quality'], categories=['low quality','Medium quality','High quality'])
wine_quality_red['wine_quality'] = wine_categorie
wine_quality_red.dtypes
```

```python
wine_quality_red
```

```python
# wine_quality_white['wine_quality'] = wine_quality_white['quality'].map(wine_quality)
wine_quality_white['wine_quality'] = wine_quality_white['quality'].apply(lambda x: 'low quality' if x<=5
                                                                       else 'Medium quality' if x<=7
                                                                       else 'High quality')
```

```python
wine_quality_white['wine_quality'] = wine_categorie
wine_quality_white.dtypes
```

```python
wine_quality_white
```

```python
# 4. Fusionner les deux datasets
wine = pd.concat([wine_quality_red, wine_quality_white], axis=0, ignore_index=True)
```
